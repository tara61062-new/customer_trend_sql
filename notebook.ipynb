```python
import pandas as pd
```

```python
df= pd.read_csv('customer_shopping_behavior.csv')
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python

df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
```

```python
--1. What is the total revenue generated by male vs female customer?
select gender, sum(purchase_amount) as revenue
	from 'customer_shopping_behavior.csv'
group by gender
```

```python
--2. Which customer used a discount but still spent more than the average purchase amount?
select customer_id, purchase_amount from 'customer_shopping_behavior.csv'
where discount_applied = 'yes' and purchase_amount > (select AVG(purchase_amount) from 'customer_shopping_behavior.csv')
```

```python
--3. Which are the top 5 products with the highest average review rating?
select item_purchased, round(avg(review_rating),2) as avg 
	from 'customer_shopping_behavior.csv'
group by item_purchased
order by avg DESC
limit 5;

-- This quesiton is really important as these items can be sold in premium prices.
```

```python
--4. Compare the average purchase amounts between standard and express shipping.
select shipping_type, round(avg(purchase_amount),2) 
	from 'customer_shopping_behavior.csv'
	where shipping_type in ('Express', 'Standard') 
	group by shipping_type

-- My conclusion - the business should invest more on Express and faster delivery options.
```

```python
--5. Do subscribed customers spend more? Compare average spend and total revenue between subscribed and non-subscriber?
select subscription_status, 
		count(customer_id), 
		round(avg(purchase_amount),2) as average_spend, 
		sum(purchase_amount) as total_revenue
	from 'customer_shopping_behavior.csv'
group by subscription_status
order by total_revenue;


-- No, un-subscribed customer have made purchased more by 0.38%.
```

```python
--6. Which 5 products have the highest percentage of purchases with discount applied?
select item_purchased, round(avg(purchase_frequency_days),2) from 'customer_shopping_behavior.csv'
where discount_applied = 'true' 
group by item_purchased
limit 5;
```
